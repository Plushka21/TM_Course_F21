```python
import numpy as np
import matplotlib.pyplot as plt

mA = 1
mB = 3
mD = 20
g = 9.8
psi = 0.6 / 100
RB = 20 / 100
RD = 20 / 100
iB = 18 / 100
rB = 0.8 * RB
JB = mB * iB ** 2
JD = mD * iB ** 2

ds = 0.0001
s = np.arange(0, 5, ds)
v = []
for d in s:
  numer = 2 * (mA * g * d - psi * mD * g * d * RB / rB * np.sin(d / rB))
  den = mA + JB / (rB ** 2) + mD * (RB / rB * np.sin(d / rB)) ** 2 + JD * (RB / (rB * RD) * np.sin(d / rB) ** 2)
  v.append(np.sqrt(numer / den))

fig, axes = plt.subplots(1, 1, figsize=(15, 5))
axes.set_xlabel("Distance, s")
axes.set_ylabel("Velocity, m/s")
axes.set_title("Plot for v_A(s)")
axes.plot(s, v, color='red')
axes.grid()
```
